```python
## palindrom olup olmadn test eden program?
# "palindrome" tersi de ayn olan liste ya da karakter dizisi(string)
text = input("Text..:")
reverse_text = text[::-1]
print(text == reverse_text and "palindrom" or "palindrom deil")
```

```python
text = input("Text..:")
liste = list(text)
reverse_text = liste[::-1]
print(liste == reverse_text and "palindrom" or "palindrom deil")
```

```python
text = input("Text..:")
liste = list(text)
reverse_text = list(reversed(liste))
print(reverse_text)
#reverse_text = ## liste[::-1]
print(liste == reverse_text and "palindrom" or "palindrom deil")
```

```python
text = input("Text..:")
liste = list(text)
copy_liste = liste.copy()
copy_liste.reverse()  ## kopyaladm ters evirdim
#reverse_text = ## liste[::-1]
print(liste == copy_liste and "palindrom" or "palindrom deil")
```
