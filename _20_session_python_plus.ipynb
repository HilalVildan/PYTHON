```python
# rumi.txt dosyasını with ... as ile açalım
# yukarıda yaptığımız işlemleri (tell, seek, read(26), readline, readlines)
# closed() dosyanın kapalı olup olmadığını test eden bir metot...
```

```python
#by Çağatay
with open("rumi.txt","r") as f:
    print(f.read(36))
    imlec = f.tell()
    print(imlec)
    icerik = f.read(imlec + 13)
    print(icerik)
    print(f.tell()) #imleç konumu
    f.seek(0)
    print(f.readline())
    print(f.closed)


print(f.closed)
```

```python
f = open(r"fishes.txt")
print("dosya adı..:", f.name)
print("Dosya kapalı mı..:",f.closed)
print("Dosya açılış amacı..:", f.mode)
f.close()
print("Dosya kapalı mı..:",f.closed)
```
